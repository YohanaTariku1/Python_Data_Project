```python
count = 0
max_name = None
max_years = 0

for name, year in years_experience.items():
  if(year > 10):
    print(f'{name} has {year} years of experience')
    count += 1

  if year > max_years:
    max_years = year
    max_name = name

print(f'Total number of applicants with 10+ years of experience are {count}.')
print(f'The most experienced applicant is {max_name} with {max_years} years of experience.')
```
