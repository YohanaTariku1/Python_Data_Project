```python
class YohanasList:
    def __init__(self):
        """Initialize an empty list ."""
        self._items = []

    def add(self, item):
        """Add an item to the list."""
        self._items.append(item)

    def __getitem__(self, index):
        """Retrieve an item by index."""
        return self._items[index]

    def __setitem__(self, index, value):
        """Set an item at a specific index."""
        self._items[index] = value

    def __repr__(self):
        """Return a string representation of the list."""
        return str(self._items)

    def __len__(self):
        """Return the number of items in the list."""
        return len(self._items)

    def __removeitem__(self, item):
        """Remove an item from the list."""
        self._items.remove(item)
```

```python
class BaseSalary:
    def __init__(self, base_salary, bonus_rate = .1, currency = '$'):
        self.base_salary = base_salary
        self.bonus_rate = bonus_rate
        self.currency = currency
        self.total_salary = base_salary * (1 + bonus_rate)
        self.bonus_salary = self.total_salary - base_salary

    def __repr__(self):
        return f'{self.currency}{self.base_salary:,.2f}'

    def show_salary(self):
        return f'{self.currency}{self.total_salary:,.2f}'

    def show_bonus(self):
        return f'{self.currency}{self.bonus_salary:,.2f}'
```

```python
salary = BaseSalary(10000)
```
