### Research Question

How have the most in-demand skills for Data Analysts in the US changed over time?

### Approach

- Aggregate skill counts by month

- Re-analyze based on percentage of total jobs

- Visualize monthly trends to show how skill popularity is shifting

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
df_DA_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_DA_pivot.loc['Total'] = df_DA_pivot.sum()

df_DA_pivot = df_DA_pivot.sort_values(by='Total', axis=1, ascending=False)

df_DA_pivot = df_DA_pivot.drop('Total')

df_DA_pivot
```

```python
DA_totals = df_DA_US.groupby('job_posted_month_no').size()

DA_totals
```

```python
df_DA_US_percent = df_DA_pivot.div(DA_totals, axis=0) * 100

df_DA_US_percent
```

```python
# changes month number to month name
df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent['job_posted_month'] = df_DA_US_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month')
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_month_no')

df_DA_US_percent
```

```python
df_plot = df_DA_US_percent.iloc[:, :5]

sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Data Analyst Job Skills Trend in the US (2023)', fontsize=16)
plt.xlabel('2023')
plt.ylabel('Likelihood in Job Posts')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
plt.gca().yaxis.set_major_formatter(PercentFormatter(decimals=0))

palette = sns.color_palette('tab10')
offsets = [0.7, -0.7, 0.4, -1.2, 0.9]

for i, skill in enumerate(df_plot.columns):
    plt.text(11.2,
             df_plot.iloc[-1, i] + offsets[i],
             skill,
             color=palette[i])

plt.show()
```
